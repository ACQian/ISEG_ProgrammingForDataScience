```python
import pandas as pd
```

```python
file="https://github.com/masterfloss/data/raw/refs/heads/main/AI_Eurostat.xlsx"
```

```python
df=pd.read_excel(file)
```

```python
df
```

```python
df = df.rename(columns={'Unnamed: 0': 'Country'})
df.head()
```

```python
df.columns
```

```python
columns_to_convert = ['USE', 'HR', 'EXPERT', 'N_LAW', 'N_ETHICS', 'N_USEFUL', 'FIN']
for col in columns_to_convert:
    df[col] = df[col].replace(':', pd.NA)
    df[col] = pd.to_numeric(df[col])

df.head()
```

```python
from numpy import average
average(df.USE.dropna())
```

```python
df_avg_by_country = df.groupby('Country')[['USE', 'HR', 'EXPERT', 'N_LAW', 'N_ETHICS', 'N_USEFUL', 'FIN']].mean()
```
