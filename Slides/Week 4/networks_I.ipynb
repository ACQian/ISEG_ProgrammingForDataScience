```python
# Import libraries
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create a network
G = nx.Graph()
```

```python
# add node to a network
G.add_node('Mary')
```

```python
# adding nodes using a list
G.add_nodes_from(['Mary', 'Steven', 'Alice','John'])
```

```python
# remove nodes
G. remove_node('Mary')
```

```python
# View nodes
G.nodes
```

```python
# remove several nodes
G.remove_nodes_from(['Mary', 'Steven'])
```

```python
# see all nodes
G.nodes
```

```python
# add several edges (list of tuples)
G.add_edges_from([('Mary', 'Steven'),('Mary', 'Alice'),('Mary','John'),('Mary','Edward')])
```

```python
# view all the edges of a network
G.edges
```

```python
# add edges 
G.add_edge('Mary','Steven')
```

```python

# remove edges
G.remove_edge('Mary','Alice')
```

```python
# number of nodes in a network
G.number_of_nodes()
```

```python
# number of edges in a network
G.number_of_edges()
```

```python
# neighbors of a node (result is a dictionary)
neighbors = G.neighbors('Alice')
neighbors
```

```python
# number of neigbors
G.degree('Mary')
```

```python
# save network
nx.write_edgelist(G, "parte1")
```

```python
# delete content of a network
G.clear()
```

```python
# read the content of a network
G = nx.read_edgelist("parte1")
```

```python
G.nodes
```

```python
#draw network
nx.draw_networkx(G)
```

```python
#draw network with labels
nx.draw_networkx(G, with_labels=True)
```

```python
# set weight to edges
G.add_edge('Mary','Steven', weight=2, distance=500)
G.add_edge('John','Mary', weight=10)
G.add_edge('Mary','Alice', weight=200)
G.add_edge('Mary','Alice', weight=300)
```

```python
# get weight from edges
G['Mary']['Alice']
```

```python
nx.draw_networkx(G, with_labels=True)
```

```python

# change the weight of an edge
G['Mary']['Steven']['weight'] = 6
```

```python
# multigraphs allows to store several proprerties
# from the same edges
# Undirected graphs with self loops and parallel edges
MG = nx.MultiGraph()
MG.add_weighted_edges_from([(1, 2, 3.0), (1, 2, 75), (2, 3, 5), (1,1,2)])
```

```python
# show edges without weight
MG.edges
```

```python
# show edge data including weight
MG.edges.data('weight', default=1)
```

```python
nx.draw_networkx(MG, with_labels=True)
```

```python
# get the weight of a edge 
MG[1][2]
```
